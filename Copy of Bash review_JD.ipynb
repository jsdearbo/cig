# Use Python to create a list.txt
Execute this code block. It's in Python. It sets up a text file for subsequent steps.

```python
samples = ['control', 'treatmentA', 'treatmentB']

file = open("list.txt", "w")
for line in samples:
    file.write("%s\n" % line)
file.close()
print('Done')
```

# Exercise #1 For loop in Bash
Write a bash script with a for loop that echoes the contents of the list.txt created in the previous step.

```python
%%bash
#!/bin/bash

# Loop through each line in list.txt and echo its contents
for line in $(cat list.txt)
do
  echo $line
done
```

# Exercise #3
Write a bash script that checks the name of the princess. If the princess is Belle, print "Lumiere." If the princess is Ariel, print "Flounder".

```python
%%bash
#!/bin/bash

# Read the contents of config.txt into an array
mapfile -t config_array < "config.txt"

# Get the name of the princess (assuming it's always on the second line)
princess="${config_array[1]}"

# Check the name of the princess and print the corresponding character
if [ "$princess" = "Belle" ]; then
    echo "Lumiere"
elif [ "$princess" = "Ariel" ]; then
    echo "Flounder"
else
    echo "Unknown Princess"
fi
```

# Exercise #4
Write a bash script that downloads the fastq file for this SRA entry https://www.ncbi.nlm.nih.gov/sra/SRX21497176[accn]

```python
%%bash
#!/bin/bash

# The above entry is too big for wget. you will need the SRA toolkit to download.
# Alternate entry selected: SRX21497196

# Download the fastq file using wget and save as SRR25774579.fastq.gz
wget -O SRR25774579.fastq.gz https://www.be-md.ncbi.nlm.nih.gov/Traces/sra-reads-be/fastq?acc=SRR25774579
```

# Exercise #5
Write a bash script that zips that fastq up

```python
%%bash
#!/bin/bash
# file is downloaded in gz format. gunzipped first.
gunzip SRR25774579.fastq.gz

# Zip the gunzipped fastq file
gzip SRR25774579.fastq
```

# Exercise #6
Write a bash script that prints each line of the fastq that starts with ">" without unzipping it. You can use the format <(file.fq.gz) in place of the unzipped file name as a process substitution.

```python
%%bash
#!/bin/bash
zcat SRR25774579.fastq.gz | grep "^>"
```

# Exercise #7
Take your command from above and instead of printing the output, pipe it to a cut command. Specify the delimiter as a space and extract the first field.

```python
%%bash
#!/bin/bash

# cut the first field of each line that starts with ">"
zcat SRR25774579.fastq.gz | grep "^>" | cut -d ' ' -f 1
```

# Exercise #8
Take your command from above and edit it so that it publishes the output to a file named "decoys.txt".

```python
%%bash
#!/bin/bash

# Print the first field of each line that starts with ">" and save it to "decoys.txt"
zcat SRR25774579.fastq.gz | grep "^>" | cut -d ' ' -f 1 > decoys.txt
```
